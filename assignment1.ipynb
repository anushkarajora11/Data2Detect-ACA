```python
# Question 1
def numberVowels(list):
    vowels = 'aeiou'
    def countVowels(word):
        return sum(1 for char in word if char in vowels)

    total = sum(countVowels(word) for word in list)
    return total
list = ["banana", "apple", "grape", "kiwi", "mango", "blueberry", "fig"]
totalNoVowels = numberVowels(list)
print("Total number of vowels:", totalNoVowels)

```

```python
# Question 2
def difference_diagonal(matrix):
  n = len(matrix)
  sum_1 = sum(matrix[i][i] for i in range(n))
  sum_2 = sum(matrix[i][n-i-1] for i in range(n))
  return abs(sum_1 - sum_2)

matrix =[ [11, 2,  4],
			    [ 4, 5,  6],
			    [10, 8, -12]]
print(difference_diagonal(matrix))


```

```python
# Question 3
def longest_sequence(lst):
    num_set = set(lst)
    longest = 0

    for num in num_set:
      if num - 1 not in num_set:
            current = num
            length = 1

            while current + 1 in num_set:
                current += 1
                length += 1

            longest = max(longest, length)

    return longest

# Example usage
lst = [10, 5, 12, 3, 55, 30, 4, 11, 2]
print(longest_sequence(lst))
```

```python
# Question 5
def factorial(num):
    if num == 0:
        return 1
    else:
        return num * factorial(num - 1)

def prime_number(num):
    if num < 2:
        return 0
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return 0
    return 1

def list_number():
    store = []
    for num in range(101):
        if prime_number(num):
            store.append(factorial(num))
    return store

final_list = list_number()
print(final_list)


```
